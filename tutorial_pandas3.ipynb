Fazendo mudanças no DataFrame

```
df.head(5)
```

```
cols = list(df.columns.values)
cols
```

```
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
# Aqui estou mudando a ordem das colunas no DataFrame
```

```
df.head(5)
```

Salvando o novo DataFrame

```
df.to_csv('modified.csv', index=False)
```

Filtrando dados

```
new_df = df.loc[(df['Type 1'] == 'Grass') & 
       (df['Type 2'] == 'Poison') &
       (df['HP'] >= 70)]
# Criando um DataFrame apenas com as condições especificadas
```

```
new_df
```

```
# Resetando o index do novo DataFrame
new_df.reset_index(drop=True, inplace=True)
new_df
```

```
df.loc[df['Name'].str.contains('Mega')]
# Estou localizando todos os nomes que contem a palavra 'Mega'
```

```
df.loc[~df['Name'].str.contains('Mega')]
# Estou localizando todos os nomes que NÃO contem a palavra 'Mega'
```

```
import re
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
# Aqui estou usando expressões Regex
# Preciso importar a biblioteca re (support for regular expressions)
```

```
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
# Estou retornando todos os pokemons que começam (^) com 'pi'
# flags=re.I significa que estou ignorando letras maiusculas
```

```
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df.head(10)
# Estou alterando todos os nomes das linhas da coluna Type 1
# De 'Fire' para 'Flamer'
```

```
df.rename(columns={'Type 1': 'Type1', 'Type 2': 'Type2'}, inplace=True)
```

```
df.head()
```

```
df.Type1.hist(figsize=(12,8), bins=50)
plt.xlabel("Tipo 1")
plt.ylabel("Número de Pokemons")
plt.title("Distribuição de Pokemons por Tipo 1")
plt.show()
# A maioria dos tipos 1 dos pokemons são do tipo aquático
```

```
df.Type2.hist(figsize=(12,8), bins=50)
plt.xlabel("Tipo 1")
plt.ylabel("Número de Pokemons")
plt.title("Distribuição de Pokemons por Tipo 1")
plt.show()
# A maioria dos tipos 2 dos pokemons são do tipo aquático
```

```
# Varicação do ataque por Tipo 1
plt.subplots(figsize=(13,8))
sns.color_palette("tab10")
sns.boxplot(x='Type1', y='Attack', data=df, linewidth=1.5)
plt.show()
# Rock, Psychic e Ground foram os tipos com maior variação
```
