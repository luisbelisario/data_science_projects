## Operador de seleção "loc"

## Selecionando com múltiplas condições

```python
data[['Idade', 'Sexo', 'Tarifa', 'Classe', 'Sobreviveu']]\
.loc[(data['Idade']>=3) & (data['Idade']<=10) & (data['Sobreviveu']==1)].head(10)
# também é possível adicionar mais de uma condição ao loc
# aqui estou selecionando todas as crianças entre 3 e 10 anos que sobreviveram
# e mostrando apenas as 10 primeiras
# \ serve para quebrar a linha
```

```python
data[['Idade', 'Sexo', 'Classe', 'Sobreviveu']]\
.loc[(data['Idade']>14) \
     & (data['Sexo']=='male') \
     & (data['Classe']==3) \
     & (data['Sobreviveu']==1)]
# aqui estou pegando apenas os homens acima de 14 anos da 3ª classe que sobreviveram
```

```python
data[['Idade', 'Sexo', 'Classe', 'Sobreviveu']]\
.loc[(data['Idade']>14) \
     & (data['Sexo']=='male') \
     & (data['Classe']==3) \
     & (data['Sobreviveu']==1)].count()
# tambem posso aplicar as formulas matematicas com o loc
# no caso fiz a contagem e constatei que apenas 29 dos homens com os atrbutos acima sobreviveram
```

<br>\B - exceto início da palavra
<br>^ - inicio de uma String
<br>$ - final de uma string
<br>[] - caracteres em colchetes
<br>[^] - caracteres não em colchetes
<br>| - ou
<br>( ) - grupo

<br>* - nenhum ou mais
<br>+ - um ou mais
<br>? - nenhum ou um
<br>{3} - número exato
<br>{3,4} - range de Números (mínimo, máximo)

## Mudando Valores Linha Por Linha - iterrows()

```python
cabine = pd.read_csv('titanic/train.csv')
```

```python
cabine['Cabin'].head()
```

```python
cabine['Cabin'].dropna()
```

```python
for i, _ in cabine[['Cabin']].dropna().iterrows():
    cabine['Cabin'].at[i] = cabine['Cabin'].at[i][0]
```

```python
cabine[['Cabin']].dropna()
```
