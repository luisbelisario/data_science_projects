```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```
x = np.arange(0,100)
y = np.arange(0,100)
```

```
x
```

```
y
```

Repare que x e y são ARRAYS UNIDIMENSIONAIS!

```
lr = LinearRegression()
```

```
lr.fit(x, y)
```

Repare que se eu eu tentar treinar o modelo com um array unidimensional aparece um erro: "Expected 2D array, got 1D array instead"

```
x.shape
```

```
y.shape
```

```
# Ao passar o -1 estou dizendo para o reshape que quero deixar aquele axis imutável
# Ao passar o 1 estou dizendo que quero criar outra dimensao, nesse caso com 1 coluna apenas
x = x.reshape(-1,1)
```

```
x.shape
```

```
# Repare que agora o treino do modelo irá funcionar
# Só preciso converter para 2D os atributos preditores
lr.fit(x, y)
```
