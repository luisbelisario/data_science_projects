```
provas_entrada = ['NU_NOTA_CN',
 'NU_NOTA_CH',
 'NU_NOTA_LC',
 'NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'
dados_sem_notas_zero = dados_sem_notas_zero[provas].dropna()
notas_entrada = dados_sem_notas_zero[provas_entrada]
notas_saida = dados_sem_notas_zero[prova_saida]
```

```
notas_entrada
```

```
# Regressão linear é muito parecido com função matemática - f(x) = y
# Onde o x é a variável que eu tenho e y é a variável que quero prever
```

```
x = notas_entrada
y = notas_saida
```

```
from sklearn.model_selection import train_test_split
SEED = 4321
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, 
                                                        random_state=SEED)
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state=SEED)
modelo.fit(x_treino, y_treino)
```

```
predicoes_matematica = modelo.predict(x_teste)
```

```
y_teste[:5]
```

```
plt.figure(figsize=(10,8))
sns.scatterplot(x=y_teste, 
                y=y_teste-predicoes_matematica)
plt.xlim((-100, 1050))
plt.ylim((-500, 1050))
plt.xlabel('Notas real das pessoas')
plt.ylabel('Valores dos erros por pessoa')
plt.show()
```

```
resultado = pd.DataFrame()
resultado['REAL'] = y_teste
resultado['PREVISAO'] = predicoes_matematica
resultado['DIFERENCA'] = resultado['REAL'] - resultado['PREVISAO']
resultado['QUADRADO_DIFERENCA'] = (resultado['REAL'] - resultado['PREVISAO'])**2
```
